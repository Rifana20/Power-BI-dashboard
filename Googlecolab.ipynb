```python
df=df[['coaster_name','Location','Manufacturer','year_introduced','latitude','longitude','Type_Main','opening_date_clean','speed_mph','height_ft','Inversions_clean','Gforce_clean']].copy()
```

```python
df['opening_date_clean']=pd.to_datetime(df['opening_date_clean'])
```

```python
df=df.rename(columns={'coaster_name':'Coaster_Name',
                      'year_introduced':'Year_Introduced',
                      'speed_mph':'Speed_mph',
                      'height_ft':'Height_ft',
                      'Inversions_clean':'Inversions',
                      'Gforce_clean':'Gforce'})
```

```python
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'opening_date_clean'])].reset_index(drop=True).copy()
```

```python
df['Year_Introduced'].value_counts()
```

```python
ax=df["Year_Introduced"].value_counts() \
  .head(10) \
  .plot(kind='bar',title='top 10 years coasters intoduced')
ax.set_xlabel('year introduced')
ax.set_ylabel('count')

```

```python
ax=df["Speed_mph"].plot(kind='hist',
                  bins=20,
                  title='coaster speed(mph)')
ax.set_xlabel('speed mph')

```

```python
ax=df["Speed_mph"].plot(kind='kde',
                  title='coaster speed')
ax.set_xlabel('speed (mph)')
```

```python
df.plot(kind='scatter',
        x='Speed_mph',
        y='Height_ft',
        title='coaster speed vs height')
plt.show()
```

```python
ax=sns.scatterplot(x='Speed_mph',
                   y='Height_ft',
                   hue="Year_Introduced",
                   data=df)
ax.set_title('coaster speed vs height')
plt.show()
```

```python
sns.pairplot(df,
             vars=['Year_Introduced','Speed_mph','Height_ft','Inversions','Gforce'],
             hue='Type_Main')
plt.show()
```

```python
df_corr=df[['Year_Introduced','Speed_mph','Height_ft','Inversions','Gforce']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr,annot=True)
```
